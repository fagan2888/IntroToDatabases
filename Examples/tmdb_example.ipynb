```python
import csv

def read_csv(fn):
    result = []
    with open(fn) as in_file:
        csvr = csv.DictReader(in_file)
        for r in csvr:
            result.append(r)
        
    return result
        
```

```python
def get_all_id_tags(l, id_col, tags_col):
    result = []
    
    for f in l:
        tmp = get_id_tag(f, id_col, tags_col)
        result.extend(tmp)
        
    return result
```

```python
%%sql
select
	a.id as film_id, a.original_title, a.overview, a.budget, a.popularity, tag_id, name as tag_name
from
	(select * from tmdbsample.films join tmdbsample.film_genres on id=film_id) as a
join
	tmdbsample.genres using(tag_id)
where
	name='Action';
```

```python
%%sql
use tmdbsample;

drop view if exists films_genres_names;

create view films_genres_names as
select
	film_id, original_title, overview, budget, popularity, tag_id,
    (select name from genres where genres.tag_id=film_genres.tag_id) as genre_name
from films join film_genres on (films.id=film_id)
```

```python
%sql select * from films_genres_names limit 10;
```

```python
%%sql
use tmdbsample;

drop view if exists films_keywords_names;

create view films_keywords_names as
select
	film_id, original_title, overview, budget, popularity, tag_id,
    (select name from keywords where keywords.tag_id=film_keywords.tag_id) as keyword_name
from films join film_keywords on (films.id=film_id)
```

```python
%sql select * from films_keywords_names limit 100;
```

```python
%%sql
drop view if exists films_tags_all;

create view films_tags_all as
select film_id, 
    films_keywords_names.tag_id as keyword_tag_id,
    keyword_name,
    films_genres_names.tag_id as genre_tag_id,
    genre_name
from
    films_genres_names join films_keywords_names
using(film_id)

    
```

```python
%sql select * from films_tags_all limit 100;
```

```python
%%sql

select
    film_id,
    (select original_title from films where films.id=film_id) as title,
    (select overview from films where films.id=film_id) as overview,
    group_concat(distinct keyword_tag_id) as all_keyword_ids,
        group_concat(distinct keyword_name) as all_keyword_names,
        group_concat(distinct genre_tag_id) as all_genre_ids,
        group_concat(distinct genre_name) as all_genre_names
from films_tags_all
where
    film_id in (select film_id from films_tags_all where genre_name='Fantasy')
    and
    film_id in (select film_id from films_tags_all where genre_name='Action')
    and
    film_id in (select film_id from films_tags_all where keyword_name='3d')
group by film_id
limit 100;

```
